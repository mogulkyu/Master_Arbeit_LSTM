**Load simulation**

```python
import os
import traci  # Import the TraCI (Traffic Controller Interface) module for interacting with SUMO

NET_FILE = "simulation/V7_landau_small_opt.net.xml"
```

**Define the binaries**

```python
# Define the path to the SUMO binaries and the network file
SUMO_BIN = os.path.join("C:/", "Program Files (x86)", "Eclipse", "Sumo", "bin", "sumo.exe")
SUMO_GUI_BIN = os.path.join("C:/", "Program Files (x86)", "Eclipse", "Sumo", "bin", "sumo-gui.exe")
```

```python

# Initialize TraCI connection to SUMO
traci.start([SUMO_BIN, "-n", NET_FILE])

trafficlights = traci.trafficlight.getIDList()  # Get the list of traffic light IDs
for trafficlight in trafficlights:
    print(trafficlight)  # Print each traffic light ID

# Close the TraCI connection when done
traci.close()
```

**Start Simulation**

```python
traci.start([SUMO_GUI_BIN, "-n", NET_FILE, "-a", "simulation/vehicles.xml,simulation/flows.xml", "--gui-settings-file", "simulation/gui_settings_Landau.xml", "--start"])


for i in range(0, 3600):
    traci.simulationStep()  # Step the simulation forward by one time unit
   
    if i % 100 == 0:
        vehicles = traci.vehicle.getIDList()
    
        print(f"Step: {i}")
        print(f"Number of vehicles: {len(vehicles)}")

 #Close the TraCI connection when done
traci.close()

#print(vectors)
```
